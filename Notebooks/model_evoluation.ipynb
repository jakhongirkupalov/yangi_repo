# Cross Validation turlari 

- K-Fold
- Stratified K-Fold
- Leave-one-out(LOO)
- Time Series Split

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

df=pd.read_csv(r"C:\Users\Jahongir\Desktop\Project_One\Data\Raw_Data\titanic.csv")
```

```python
df.head()
```

```python
df = df[["Survived", "Pclass", "Sex", "Age", "Fare"]]

df["Age"].fillna(df["Age"].median(),inplace=True)
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
```

```python
x = df.drop("Survived", axis=1)
y = df["Survived"]
```

# KFOLD

```python
kfold = KFold(
    n_splits=5,  # K=5 Tavsiya qilinadi 5dan 10 tagacha
    shuffle=True,  # datasetni aralashtirish
    random_state=42
)
```

```python
pipeline = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000))
])
```

```python
x["Age"] = x["Age"].fillna(x["Age"].median())
```

```python
x.isnull().sum()
```

# CROSS VALIDATION yordamida baaaholash

```python
scores = cross_val_score(
    estimator=pipeline,
    X=x,
    y=y,
    cv=kfold,
    scoring="accuracy"
) 
```

```python
print("Har bir fold accuracy natijasi:", scores)
print("O'rtacha accuracy:", scores.mean())
print("Standart og'ish:", scores.std())
```

# Stratified K-Fold Validation

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = cross_val_score(
    pipeline,
    x,
    y,
    cv=skf,
    scoring= "accuracy"
)
scores.mean()
```
